```python
#corr = df_hnu[df_hnu.columns[~df_hnu.columns.str.contains("CH")]].corr()
corr = df_hnu.corr()
```

```python
df_smpl = df_hnu[df_hnu.columns[~df_hnu.columns.str.contains("CH")]]
corr = df_smpl.corr()
```

```python
np.fill_diagonal(corr.values, np.nan)
corr
```

```python
valor_umbral = 0.9
indice_i, indice_j = np.where(np.abs(corr.values) > valor_umbral)
asociacion_muy_alta = [(corr.index[i], corr.columns[j], corr.values[i,j]) for i, j in zip(indice_i, indice_j)]
asociacion_muy_alta
```

```python
valor_max = 0.9
valor_min = 0.7
indice_i, indice_j = np.where((np.abs(corr.values) >= valor_min) & (np.abs(corr.values) < valor_max))
asociacion_alta = [(corr.index[i], corr.columns[j]) for i, j in zip(indice_i, indice_j)]
asociacion_alta
```

```python
valor_max = 0.9
valor_min = 0.7
indice_i, indice_j = np.where((np.abs(corr.values) >= valor_min) & (np.abs(corr.values) < valor_max))
indices = [(i,j) if i<=j else (j,i) for i, j in zip(indice_i, indice_j)] # ordenar pares de indices i <= j
indices = list(set(indices)) # eliminar duplicados
asociacion_alta = [(corr.index[i], corr.columns[j]) for i, j in indices]
asociacion_alta
```

```python
valor_max = 0.7
valor_min = 0.5
indice_i, indice_j = np.where((np.abs(corr.values) >= valor_min) & (np.abs(corr.values) < valor_max))
indices = [(i,j) if i<j else (j,i) for i, j in zip(indice_i, indice_j)]
indices = list(set(indices))
asociacion_moderada = [(corr.index[i], corr.columns[j]) for i, j in indices]
asociacion_moderada
```

```python
valor_min = 0.9
indice_i, indice_j = np.where(np.abs(corr.values) >= valor_min)
indices = [(i,j) if i<j else (j,i) for i, j in zip(indice_i, indice_j)]
indices = list(set(indices))
asociacion_muy_alta = [(corr.values[i,j], corr.index[i], corr.columns[j]) for i, j in indices]
asociacion_muy_alta
```
